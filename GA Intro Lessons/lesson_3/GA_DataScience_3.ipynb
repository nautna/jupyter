##### (3) Filtering / Slicing Data

```python
zillow_df.columns # another attribute; returns index of column names
```

```python
# DataFrame consists of three objects: index, columns and values
```

```python
zillow_df[['Address', 'Longitude', 'Latitude']] # slicing to get data (filter)
```

```python
# why 2 brackets after _df[[]]? because you are accessing 2 dimensional data. Values are stored in 1 dimensional series
```

```python
columns_to_select = ['Address', 'Longitude', 'Latitude']
zillow_df[columns_to_select].iloc[10:25] # choosing specific rows of interest, .iloc filters by row
```

```python
# for consistency with database, get rid of default index and utilize ID from DataFrame
new_zillow_df = zillow_df.set_index('ID')
new_zillow_df.head()
```

```python
new_zillow_df.iloc[[0, 3]] # 'i' in .iloc refers to index. must use index location (range(len(dataframe)-1))
```

```python
new_zillow_df.loc[[15063471, 15064044], ['Address', 'SalePrice']] 
# filters by rows by default, but with brackets, accepts list of rows and/or column by name
```
