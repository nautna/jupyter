```python
monthly_excel = {
    'January': 'jan.xlsx',
    'February': 'feb.xlsx',
    'March': 'mar.xlsx',
    'April': 'apr.xlsx',
    'May': 'may.xlsx'
}

#dictionary of months and file names to be run later on
```

```python
excel_agg = {} #loading the files into an 'aggregate' dictionary
for month in monthly_excel:
    excel_agg['{}'.format(month)] = pd.ExcelFile(r'F:\Desktop\SpecialtyTraining\Eco\FSearch\NOTM Impact\{}'.format(monthly_excel[month]))
```

```python
read_agg = {} #reading all files into dictionary
for month in excel_agg:
    read_agg['{}'.format(month)] = excel_agg[month].parse('FastSear')
```

We have the aggregate data in the above dictionary. Will try to clean this to reflect the relevant inquiries/remove duplicates.

```python
bucket_5 = {} #dictionary of GDA resolved tickets tagged with bucket 5
for month in read_agg: #for loop to give each month's tickets with bucket 5 ticket (month: df)
    bucket_5['{}'.format(month)] = read_agg[month][read_agg[month]['Tags Desc'] == 'Data>Bucket>115-GDA>5-Eco Stats']

total = {} #for total tickets resolved by GDA to track BBGID since ADSK FS does not output multiple tags per ticket
for month in read_agg: #for loop to give each month's tickets resolved by GDA with BBGID provided (month: df)
    total['{}'.format(month)] = read_agg[month][read_agg[month]['Tags Desc'] != 'Data>Bucket>115-GDA>5-Eco Stats']
```

```python
bucket_5_renamed = {} #initializing dictionaries to hold data with renamed columns
total_renamed = {}

for month in bucket_5:
    bucket_5_renamed[month] = bucket_5[month].rename(columns={'Tags Desc': 'Bucket Tag'}) #renaming columns
for month in total:
    total_renamed[month] = total[month].rename(columns={'Tag Value': 'BBGID Value'}) #renaming columns of interest
```

```python
bucket_5_renamed['February'].columns
```

```python
columns_to_keep_5 = ['ADSK Number', 'Message Number', 'Creation Time (UTC)', 'Action Time (UTC)', 'Resolve Time (UTC)', 'Bucket',
                   'Bucket Tag', 'Inquiry Rep Region']
columns_to_keep_tot = ['ADSK Number', 'Message Number', 'Creation Time (UTC)', 'Action Time (UTC)', 'Resolve Time (UTC)', 'Bucket',
                   'BBGID Value', 'Inquiry Rep Region', 'Inquiry Rep Desc']
```

```python
final_5 = {}  #we are dropping columns that are not of interest here. will join frames below
final_tot = {}

for month in bucket_5_renamed:
    final_5[month] = bucket_5_renamed[month][columns_to_keep_5]
for month in final_tot:
    final_tot[month] = total_renamed[month][columns_to_keep_tot]
```

```python
agg_5 = pd.concat(final_5.values(), ignore_index=True) #adding all months to one dataframe
agg_tot = pd.concat(final_tot.values(), ignore_index=True)
```
